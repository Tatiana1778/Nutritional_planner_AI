```python
import pandas as pd
```

```python
# Sample Data (You can replace this with your actual data)
dataset = pd.read_csv("/content/sample_data/food_dataset.csv", delimiter=';')
dataset.head(15)
```

```python
# Group the data by 'Food category' and 'Type of food', and count occurrences
category_subcategory_counts = dataset.groupby(['Food category', 'Type of food']).size().reset_index(name='Count')
category_subcategory_counts = category_subcategory_counts.sort_values(by='Food category')
# Print the categories with their subcategories and counts
print("Food Categories and Subcategories:\n")
for category in category_subcategory_counts['Food category'].unique():
    print(f"\n{category}:")
    subcategories = category_subcategory_counts[category_subcategory_counts['Food category'] == category]
    for _, row in subcategories.iterrows():
        print(f"  - {row['Type of food']} (Count: {row['Count']})")
```

```python
import random
# Function to pick one random type of food from specified category
def pick_random_from_specified_categories(dataset, specified_categories):
    random_foods = {}
    for category in specified_categories:
        if category in dataset['Food category'].unique():  # Check if the category exists in the dataset
            subcategory = dataset[dataset['Food category'] == category]['Type of food']
            random_foods[category] = random.choice(subcategory.tolist())  # Randomly pick a food type
        else:
            random_foods[category] = "Category not found"  # Handle non-existing categories
    return random_foods

# Specify the categories you want to pick from
specified_categories = ['vegetable', 'seafood', 'grain']

# Pick random foods
random_picks = pick_random_from_specified_categories(dataset, specified_categories)

# Display the result
print("Randomly selected food from specified categories:")
for category, food in random_picks.items():
    print(f"{category}: {food}")
```

```python
import random

# Function to pick one random type of food from specified categories
def pick_random_from_specified_categories_with_nutrition(dataset, specified_categories, exclude_foods=None):
    # Specify the nutrition details to display
    nutrition_columns = [
        'Caloric Value (kcal)', 'Fat (g)', 'Saturated Fats (g)', 'Monounsaturated Fats (g)',
        'Polyunsaturated Fats (g)', 'Carbohydrates (g)', 'Sugars (g)', 'Protein (g)',
        'Dietary Fiber (g)', 'Cholesterol (mg)', 'Sodium (mg)', 'Water (g)', 'Vitamin A (IU)',
        'Vitamin B1 (mg)', 'Vitamin B11 (mcg)', 'Vitamin B12 (mcg)', 'Vitamin B2 (mg)',
        'Vitamin B3 (mg)', 'Vitamin B5 (mg)', 'Vitamin B6', 'Vitamin C', 'Vitamin D',
        'Vitamin E', 'Vitamin K', 'Calcium', 'Copper', 'Iron', 'Magnesium', 'Manganese',
        'Phosphorus', 'Potassium', 'Selenium', 'Zinc', 'Nutrition Density'
    ]
    key_nutrients = ['Carbohydrates (g)', 'Protein (g)', 'Fat (g)']
    random_foods = {}

    # Convert all food categories and excluded foods to lowercase for consistency
    dataset['Food category'] = dataset['Food category'].str.lower()
    dataset['Type of food'] = dataset['Type of food'].str.lower()
    exclude_foods = [food.lower() for food in (exclude_foods or [])]

    for category in specified_categories:
        category_lower = category.lower()  # Ensure case-insensitive comparison

        if category_lower in dataset['Food category'].unique():  # Check if the category exists
            subcategory = dataset[dataset['Food category'] == category_lower]

            # Filter out foods that are in the exclusion list
            valid_foods = subcategory[~subcategory['Type of food'].isin(exclude_foods)]
            excluded_foods = subcategory[subcategory['Type of food'].isin(exclude_foods)]

            if not valid_foods.empty:  # Check if there are any valid foods left
                selected_food = valid_foods.sample(n=1).iloc[0]  # Randomly pick one food
                nutrition_values = selected_food[nutrition_columns].to_dict()  # Get nutrition details
                nutrition_values['Type of food'] = selected_food['Type of food']  # Add food name
                random_foods[category] = nutrition_values  # Store food details
            elif not excluded_foods.empty:  # Handle the case when valid foods are unavailable
                # Identify the excluded food with the highest value in key nutrients
                excluded_food = excluded_foods.iloc[0]  # Take the first excluded food for simplicity
                highest_nutrient = excluded_food[key_nutrients].idxmax()

                # Find an alternative food with the highest value in the same nutrient
                alternative_food = subcategory.loc[subcategory[highest_nutrient].idxmax()]
                nutrition_values = alternative_food[nutrition_columns].to_dict()
                nutrition_values['Type of food'] = alternative_food['Type of food']
                nutrition_values['Replaced food due to exclusions'] = excluded_food['Type of food']
                random_foods[category] = nutrition_values
            else:
                random_foods[category] = {"message": "No valid foods available in this category after exclusions"}
        else:
            random_foods[category] = {"message": "Category not found"}  # Handle non-existing categories

    return random_foods

# Specify the categories you want to pick from
specified_categories = ['vegetable', 'meat', 'grain']

# List of foods that cannot be eaten (example)
exclude_foods = ['spinach', 'lamb', 'oats']

# Pick random foods
random_picks = pick_random_from_specified_categories_with_nutrition(dataset, specified_categories, exclude_foods)

# Display the result
print("Randomly selected food from specified categories with nutrition values:\n")
for category, details in random_picks.items():
    print(f"{category}:")
    if "message" in details:
        print(f"  {details['message']}")
    else:
        print(f"  Type of Food: {details['Type of food'].capitalize()}")
        if 'Replaced food due to exclusions' in details:
            print(f"  Replaced Food: {details['Replaced food due to exclusions'].capitalize()}")
        print("  Nutrition Values:")
        for key, value in details.items():
            if key not in ['Type of food', 'Replaced food due to exclusions']:  # Skip non-nutritional keys
                print(f"    {key}: {value}")
```

```python
import random

# AI suggestion function
def ai_recommendation(category, excluded_foods, dataset):
    """
    Suggest a food alternative from the same or a related category, excluding specified foods.
    """
    dataset['Food category'] = dataset['Food category'].str.lower()
    category_data = dataset[dataset['Food category'] == category.lower()]['Type of food']

    # Filter out excluded foods
    valid_foods = [food for food in category_data.tolist() if food.lower() not in excluded_foods]

    if valid_foods:
        return random.choice(valid_foods)  # Suggest a random valid food
    else:
        # If no valid foods in the same category, suggest a food from any other category
        all_foods = [food for food in dataset['Type of food'].tolist() if food.lower() not in excluded_foods]
        return random.choice(all_foods) if all_foods else "No valid alternatives available"

# Function to pick one random type of food from specified category
def pick_random_from_specified_categories_with_ai(dataset, specified_categories, exclude_foods=None):
    random_foods = {}
    ai_suggestions = {}

    # Convert all food categories to lowercase to ensure case-insensitive comparison
    dataset['Food category'] = dataset['Food category'].str.lower()

    # If no foods to exclude, initialize the list as an empty list
    if exclude_foods is None:
        exclude_foods = []

    for category in specified_categories:
        category_lower = category.lower()  # Make the specified category lowercase for consistency

        if category_lower in dataset['Food category'].unique():  # Check if the category exists
            subcategory = dataset[dataset['Food category'] == category_lower]['Type of food']

            # Filter out foods that are in the exclusion list
            valid_foods = [food for food in subcategory.tolist() if food.lower() not in exclude_foods]

            if valid_foods:  # Check if there are any valid foods left after exclusion
                selected_food = random.choice(valid_foods)  # Randomly pick a food type
                random_foods[category] = selected_food
            else:
                random_foods[category] = "No valid foods available in this category after exclusions"
                # AI suggestion
                ai_suggestions[category] = ai_recommendation(category, exclude_foods, dataset)
        else:
            random_foods[category] = "Category not found"  # Handle non-existing categories

    return random_foods, ai_suggestions

# Specify the categories you want to pick from
specified_categories = ['vegetable', 'seafood', 'grain']

# List of foods that cannot be eaten (example)
exclude_foods = ['spinach', 'salmon', 'oats']

# Pick random foods and get AI suggestions
random_picks, ai_suggestions = pick_random_from_specified_categories_with_ai(dataset, specified_categories, exclude_foods)

# Display the result
print("Randomly selected food from specified categories (excluding certain foods):")
for category, food in random_picks.items():
    print(f"{category}: {food}")
    if category in ai_suggestions and food.startswith("No valid"):
        print(f"  AI Suggestion: {ai_suggestions[category]}")
```
